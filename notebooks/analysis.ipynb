```python
# Анализ покупательского поведения и влияния подписки

# Цель проекта — проанализировать данные о покупках и проверить,
# влияет ли наличие подписки на средний чек и поведение покупателей.

import pandas as pd

print('ок')

df = pd.read_csv('C:/Users/Admin/PycharmProjects/analytics_project/data/shopping_behavior_updated.csv', encoding='windows-1251')

# Общая информация по данным
# df.info()

# Посчитает пропуски по каждой колонке
df.isna().sum()

# Проверяет есть ли пропуски вообще
# df.isna().sum().sum()

# Вывод №1: «На этапе разведочного анализа явных пропусков не выявлено»

df = df.rename(columns= {
    'Customer ID': 'id',
    'Age': 'age',
    'Gender': 'gender',
    'Item Purchased': 'purchased_item',
    'Category': 'category',
    'Purchase Amount (USD)': 'price',
    'Location': 'location',
    'Size': 'size',
    'Color': 'color',
    'Season': 'season',
    'Review Rating': 'rating',
    'Subscription Status': 'subscription',
    'Shipping Type': 'shipping',
    'Discount Applied': 'discount_applied',
    'Promo Code Used': 'promo_used',
    'Previous Purchases': 'purchases_amount',
    'Payment Method': 'payment_method',
    'Frequency of Purchases': 'purchase_frequency'
})

# print(df.head())

# Вопрос 1
avg_check = df.groupby('subscription')['price'].mean()

print(avg_check)

# 2 вопрос
basic_metrics= df.groupby('subscription').agg(
    orders_count = ('id', 'count'),
    total_revenue = ('price', 'sum'),
    avg_check = ('price', 'mean')
)

print(basic_metrics)

# Вывод 2:
# Средний чек у клиентов с подпиской и без подписки практически не отличается.
# Общее количество заказов и суммарная выручка выше у клиентов без подписки, что может быть связано с большим числом таких клиентов в датасете, а не с более высокой индивидуальной частотой покупок.
# На основе текущих агрегатов нельзя утверждать, что подписка влияет на частоту покупок одного клиента.

client_behaviour = df.groupby('subscription').agg(
    customers=('id', 'nunique'),
    orders=('id', 'count'),
    revenue=('price', 'sum')
)

print(client_behaviour)

# Датасет содержит по одной покупке на клиента, поэтому анализ частоты повторных покупок ограничен.
```
